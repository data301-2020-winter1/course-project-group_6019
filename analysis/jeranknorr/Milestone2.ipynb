```python
import numpy as np
import pandas as pd
import scripts as sc

def load_and_process(Madden21_Clean):
    df= (
        pd.read_csv('../../data/raw/madden21_ratings.csv')
        .dropna()
    )
   
    df2= (    
            df.filter(['Team', 'Full Name', 'Overall Rating', 'Position'])
)
    return df2
```

```python
dfQB = (
        df.loc[lambda x: x['Position'].str.contains('QB')]
        .groupby('Team')['Overall Rating'].max()
        .reset_index()
        .sort_values('Overall Rating', ascending=False).reset_index(drop=True)
        .rename(columns={'Overall Rating': 'Top Quaterback'})
)
 #Max value for Each quater back
```

```python
dfTE = (
        df.loc[lambda x: x['Position'].str.contains('TE')]
        .groupby('Team')['Overall Rating'].max()
        .reset_index()
        .sort_values('Overall Rating', ascending=False).reset_index(drop=True)
        .rename(columns={'Overall Rating': 'Top Tight End'})
)
 #Max value for Each Tight end
```

```python
df10 = pd.merge(dfQB,dfWR).merge(dfTE).merge(dfHB)
df10['Offence_Weighted_Average']=(df10['Top Quaterback']*.4)+(df10['Top Wide Receiver']*.2)+(df10['Top Tight End']*.2)+(df10['Top Half Back']*.2)
df10.sort_values('Offence_Weighted_Average', ascending=False).reset_index(drop=True)
```
